```python
# Draw bounding box to check

import os
import cv2

class_name = [
    'Jeff_Bezos',
    'Bill_Gates',
    'Elon_Musk',
    'Larry_Page',
    'Mark_Zuckerberg'
]

for current_dir, dirs, files in os.walk('label'):
    for f in files:
        if not f.endswith('.txt'):
            continue
        fname = f[:-4]
        print(fname)
        img = cv2.imread(f'dataset/{fname}.jpg')
        height, width, channels = img.shape
        with open(f'label/{f}', 'r') as f2:
            for line in f2:
                a = [float(x) for x in line.strip().split()]
                a[0] = int(a[0])
                print(a)
                x = int(a[1] * width)
                y = int(a[2] * height)
                w = int(a[3] * width)
                h = int(a[4] * height)
                x1 = int(x - w / 2)
                x2 = int(x + w / 2)
                y1 = int(y - h / 2)
                y2 = int(y + h / 2)
                print(x1, y2, x2, y2)
                cv2.rectangle(img, (x1, y1), (x2, y2), (36,255,12), 2)
                fontScale = min(width,height) / 500
                fontScale = 1
                cv2.putText(img, class_name[a[0]], (x1, y1 + 30), cv2.FONT_HERSHEY_SIMPLEX, fontScale, (36,255,12), 2)
        cv2.imwrite(f'process/{fname}.jpg', img)
```

```python
import os

a = ['bezos', 'gates', 'musk', 'page', 'zuckerberg']
b = [0] * len(a)
c = [[], [], [], [], []]

for current_dir, dirs, files in os.walk('img'):
    # Going through all files
    for f in files:
        for i, x in enumerate(a):
            if x in f:
                b[i] += 1
                c[i].append(f'{f}')
                break
for i in range(len(a)):
    print(f'{a[i]} = {b[i]}')

f2 = open('train.txt', 'w')
f3 = open('test.txt', 'w')

for i in range(len(c)):
    p_test = c[i][:int(len(c[i]) * 0.2)]
    p = c[i][int(len(c[i]) * 0.2):]
    for e in p:
        print(f'dataset/{e}', file=f2)
    for e in p_test:
        print(f'dataset/{e}', file=f3)
f2.close()
f3.close()
```
